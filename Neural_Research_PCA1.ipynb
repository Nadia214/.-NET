```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the dataset
file_path = 'cleaned_data.csv'
data = pd.read_csv(file_path)

# Load dataset
dataset = pd.read_csv("cleaned_data.csv")
array = dataset.values
X = array[:,0:8]  # Features
Y = array[:,7]    # Target (classification)

data_numeric=X

# Drop non-numeric columns and handle missing values
#data_numeric = data.drop(columns=['case']).dropna()

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_numeric)

# Apply PCA to reduce to 2 components
pca = PCA(n_components=8)
data_pca = pca.fit_transform(data_scaled)

# Create a DataFrame with the PCA results
pca_columns = ['PC1', 'PC2', 'PC3', 'PC4','PC5', 'PC6', 'PC7', 'PC8']
data_pca_df = pd.DataFrame(data_pca, columns=pca_columns)

# Save the transformed data to a new CSV
output_path = 'transformed_dataset.csv'
data_pca_df.to_csv(output_path, index=False)

print(f"Transformed dataset saved to {output_path}")
```
